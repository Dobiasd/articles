```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from matplotlib.ticker import MaxNLocator

a, b = 2, 5
scale = 30

def add_measure(pos, name):
    plt.axvline(pos, color='k', linestyle='dashed', linewidth=2)
    _, max_ = plt.ylim()
    plt.text(pos + pos/10,
             max_ - max_/10,
             '{}: {:.2f}'.format(name, pos))

def get_beta_samples():
    x = np.linspace(0, 1, 50)
    y = stats.beta.rvs(a, b, size=10000, scale=scale)
    return x, y

def plot_beta(x, y):
    plt.clf()
    a, b = 2, 5
    sns.distplot(y, hist=True, kde=False,
                 bins=scale*x,
                 hist_kws=dict(edgecolor="k", linewidth=2))
    

x, y = get_beta_samples()
plot_beta(x, y)
#add_measure(y.mean(), 'mean')
plt.show()

#plot_beta(x, y)
#add_measure(np.median(y), 'median')
#plt.show()

#plot_beta(x, y)
#add_measure(stats.mode(y)[0][0], 'mode')
#plt.show()
```

```python
def plot_beta_pdf():
    plt.clf()
    x = np.linspace(0, scale, 1000)
    y = stats.beta.pdf(x, a, b, scale=scale)
    plt.plot(x, y, 'blue', lw=2, alpha=0.7)
    plt.ylim(0)
    #plt.grid()
    
def plot_beta_cdf():
    plt.clf()
    x = np.linspace(0, scale, 1000)
    y = stats.beta.cdf(x, a, b, scale=scale)
    plt.plot(x, y, 'blue', lw=2, alpha=0.7)
    plt.yticks(np.arange(0, 1.1, step=0.1))
    plt.ylim(0)
    #plt.grid()

plot_beta_pdf()
plt.show()

plot_beta_pdf()
add_measure(stats.beta.mean(a, b, scale=scale), 'mean')
plt.show()

plot_beta_pdf()
add_measure(stats.beta.median(a, b, scale=scale), 'median')
plt.show()

plot_beta_cdf()
plt.show()

plot_beta_cdf()
add_measure(stats.beta.median(a, b, scale=scale), 'median')
plt.grid()
plt.show()
```
